```python
#ouliers removal


# Select only numeric columns
numeric_cols = dataset.select_dtypes(include=['float64', 'int64']).columns
print(numeric_cols)
```

```python
# Function to detect outliers using IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    return df[(df[column] < lower_limit) | (df[column] > upper_limit)]
```

```python
for col in numeric_cols:
    outliers = detect_outliers(dataset, col)
    print(f"{col}: {outliers.shape[0]} outliers")
```

```python
# Create a copy to preserve original data
cleaned_dataset = dataset.copy()

for col in numeric_cols:
    Q1 = cleaned_dataset[col].quantile(0.25)
    Q3 = cleaned_dataset[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    # Remove outliers
    cleaned_dataset = cleaned_dataset[
        (cleaned_dataset[col] >= lower_limit) & (cleaned_dataset[col] <= upper_limit)
    ]

print("✅ Dataset shape before:", dataset.shape)
print("✅ Dataset shape after removing outliers:", cleaned_dataset.shape)
```
